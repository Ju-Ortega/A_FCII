```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funciÃ³n original sin ruido
def f(x):
    return np.sin(x)

# FunciÃ³n con ruido aÃ±adido
def f_with_noise(x, noise_level=0.01):
    noise = noise_level * np.random.randn(*x.shape)
    return np.sin(x) + noise

# Derivada analÃ­tica exacta de la funciÃ³n
def f_prime_exact(x):
    return np.cos(x)

# Derivada adelantada
def forward_diff(f, x, h):
    return (f(x + h) - f(x)) / h

# Derivada retrasada
def backward_diff(f, x, h):
    return (f(x) - f(x - h)) / h

# Derivada centrada
def centered_diff(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Valores para la comparaciÃ³n
x = np.linspace(0, 2 * np.pi, 100)
h = 0.1  # TamaÃ±o de paso

# FunciÃ³n con ruido
f_noisy = f_with_noise(x)

# CÃ¡lculo de las derivadas con ruido
f_prime_forward_noisy = forward_diff(f_with_noise, x, h)
f_prime_backward_noisy = backward_diff(f_with_noise, x, h)
f_prime_centered_noisy = centered_diff(f_with_noise, x, h)

# Derivada exacta (sin ruido)
f_prime_true = f_prime_exact(x)

# Graficar los resultados
plt.figure(figsize=(14, 8))

# GrÃ¡fico de la funciÃ³n con ruido
plt.subplot(2, 1, 1)
plt.plot(x, f(x), label='FunciÃ³n Original', color='black')
plt.plot(x, f_noisy, label='FunciÃ³n con Ruido', color='orange')
plt.legend()
plt.title('FunciÃ³n Original y FunciÃ³n con Ruido')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# GrÃ¡fico de las derivadas
plt.subplot(2, 1, 2)
plt.plot(x, f_prime_true, label='Derivada Exacta', color='black', linestyle='--')
plt.plot(x, f_prime_forward_noisy, label='Derivada Adelantada con Ruido', color='blue')
plt.plot(x, f_prime_backward_noisy, label='Derivada Retrasada con Ruido', color='red')
plt.plot(x, f_prime_centered_noisy, label='Derivada Centrada con Ruido', color='green')
plt.legend()
plt.title('ComparaciÃ³n de MÃ©todos de DerivaciÃ³n con Ruido')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# CÃ³digo Ejercicio 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las funciones y sus derivadas exactas
def f_sin(x):
    return np.sin(x)

def f_prime_sin(x):
    return np.cos(x)

def f_exp(x):
    return np.exp(x)

def f_prime_exp(x):
    return np.exp(x)

def f_ln(x):
    return np.log(x)

def f_prime_ln(x):
    return 1/x

def f_square(x):
    return x**2

def f_prime_square(x):
    return 2*x

# Derivada adelantada
def forward_diff(f, x, h):
    return (f(x + h) - f(x)) / h

# Derivada retrasada
def backward_diff(f, x, h):
    return (f(x) - f(x - h)) / h

# Derivada centrada
def centered_diff(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Valores para la comparaciÃ³n
x = np.linspace(0.1, 2 * np.pi, 100)  # Se evita x=0 para evitar problemas con ln(x)
h = 0.1  # TamaÃ±o de paso

# Funciones a comparar
funciones = [
    (f_sin, f_prime_sin, "sin(x)"),
    (f_exp, f_prime_exp, "exp(x)"),
    (f_ln, f_prime_ln, "ln(x)"),
    (f_square, f_prime_square, "x^2")
]

plt.figure(figsize=(14, 12))

for i, (f, f_prime, nombre) in enumerate(funciones, 1):
    f_prime_forward = forward_diff(f, x, h)
    f_prime_backward = backward_diff(f, x, h)
    f_prime_centered = centered_diff(f, x, h)
    f_prime_true = f_prime(x)

    # Graficar los resultados
    plt.subplot(2, 2, i)
    plt.plot(x, f_prime_true, label='Derivada Exacta', color='black', linestyle='--')
    plt.plot(x, f_prime_forward, label='Derivada Adelantada', color='blue')
    plt.plot(x, f_prime_backward, label='Derivada Retrasada', color='red')
    plt.plot(x, f_prime_centered, label='Derivada Centrada', color='green')
    plt.legend()
    plt.title(f"ComparaciÃ³n de MÃ©todos para {nombre}")
    plt.xlabel('x')
    plt.ylabel("f'(x)")
    plt.grid(True)

plt.tight_layout()
plt.show()
```
