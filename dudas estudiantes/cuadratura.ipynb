# Derivación del error de la Regla Trapezoidal utilizando la Serie de Taylor

## Paso 1: Expansión en Serie de Taylor alrededor del punto $a$

Partimos expandiendo la función $f(x)$ en una serie de Taylor alrededor del punto $a$. La expansión de Taylor de $f(x)$ cerca de $a$ es:

$$
f(x) = f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f'''(a)}{3!}(x - a)^3 + \dots
$$

Queremos evaluar esta expansión en $x = b$. Por lo tanto, sustituimos $x = b$ en la serie de Taylor:

$$
f(b) = f(a) + f'(a)(b - a) + \frac{f''(a)}{2!}(b - a)^2 + \frac{f'''(a)}{3!}(b - a)^3 + \dots
$$

---

## Paso 2: Aplicación de la Regla Trapezoidal

La regla trapezoidal aproxima la integral definida de $f(x)$ en el intervalo $[a, b]$ como:

$$
\int_a^b f(x) \, dx \approx \frac{1}{2}(b - a) \left[ f(a) + f(b) \right].
$$

Sustituyendo la expansión de $f(b)$ en esta fórmula, tenemos:

$$
\frac{1}{2}(b - a) \left[ f(a) + \left( f(a) + f'(a)(b - a) + \frac{f''(a)}{2!}(b - a)^2 + \dots \right) \right].
$$

Simplificando la expresión:

$$
\frac{1}{2}(b - a) \left[ 2 f(a) + f'(a)(b - a) + \frac{f''(a)}{2!}(b - a)^2 + \dots \right].
$$

Distribuimos el factor $\frac{1}{2}(b - a)$:

$$
(b - a) f(a) + \frac{(b - a)^2}{2} f'(a) + \frac{(b - a)^3}{12} f''(a) + \dots
$$

---

## Paso 3: Integral Exacta de $f(x)$

La integral exacta se obtiene integrando término a término la serie de Taylor:

$$
\int_a^b f(x) \, dx = \int_a^b \left[ f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \dots \right] \, dx.
$$

Evaluamos esta integral para cada término:

$$
\int_a^b f(x) \, dx = (b - a) f(a) + \frac{(b - a)^2}{2} f'(a) + \frac{(b - a)^3}{6} f''(a) + \dots
$$

---

## Paso 4: Comparación entre la Regla Trapezoidal y la Integral Exacta

Ahora comparamos la aproximación de la regla trapezoidal con la integral exacta.

- La **regla trapezoidal** nos dio:
  $$
  (b - a) f(a) + \frac{(b - a)^2}{2} f'(a) + \frac{(b - a)^3}{12} f''(a) + \dots
  $$

- La **integral exacta** es:
  $$
  (b - a) f(a) + \frac{(b - a)^2}{2} f'(a) + \frac{(b - a)^3}{6} f''(a) + \dots
  $$

---

## Paso 5: Derivación del Error

El **error** de la aproximación se obtiene comparando los términos correspondientes en ambas expresiones. El término relevante para el error es:

$$
\frac{(b - a)^3}{6} f''(a) - \frac{(b - a)^3}{12} f''(a).
$$

Esto nos da:

$$
E_T = \frac{(b - a)^3}{12} f''(\xi),
$$

donde $\xi \in (a, b)$ es un punto intermedio (por el teorema del valor medio).
